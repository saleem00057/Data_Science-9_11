## Matplotlib

Matplotlib is a popular Python library used for creating visualizations, like graphs and charts. It helps you turn your data into pictures that are easier to understand and analyze. 
- What it does: Matplotlib allows you to make different types of plots, like line graphs, bar charts, scatter plots, and more. These visualizations help you see patterns or trends in your data.

- How it works: You give Matplotlib data (numbers, lists, arrays) and tell it what kind of graph you want to make. Matplotlib then draws the graph and shows it on your screen.

- Why it's useful: Instead of just reading raw numbers or text, you can see them in a graph or chart, which is much easier to understand, especially for large amounts of data.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]  # X-axis values
y = [2, 4, 6, 8]  # Y-axis values

# Plotting
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('My First Plot')

# Show the graph
plt.show()
```

## Line plot

```python
import matplotlib.pyplot as plt

# Data for the plot
x = [1, 2, 3, 4, 5]  # X-axis values (e.g., days)
y = [2, 4, 6, 8, 10]  # Y-axis values (e.g., sales)

# Create a line plot
plt.plot(x, y, label='Sales Over Days', color='blue', marker='o')

# Add labels and title
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Line Plot Example')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

## Key Features You Can Add
- Line Style: Make the line dashed, dotted, etc.
- Color: Change the line color
- Markers: Add markers to highlight points.

```python
plt.plot(x, y, linestyle='--')  # Dashed line
```

```python
plt.plot(x, y, color='red')  # Red line
```

```python
plt.plot(x, y, marker='o')  # Circles at points
```

## When to Use a Bar Chart
- To compare different categories or groups.
- To show data like counts, frequencies, or percentages.

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']  # Categories
values = [10, 15, 7, 12]           # Values for each category

# Create the bar chart
plt.bar(categories, values, color='skyblue')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Show the plot
plt.show()
```

### Customizing a Bar Chart
- Horizontal Bars:
- Bar Colors:
- Add Grid:
- Bar Width:

```python
plt.barh(categories, values, color='green')  # Horizontal bar chart
```

```python
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
```

```python
plt.grid(axis='y')  # Adds grid lines to Y-axis
```

```python
plt.bar(categories, values, width=0.5)  # Change bar width
```

### When to Use a Scatter Plot?
- To identify patterns, trends, or relationships between two variables.
- To detect clusters or outliers in data.

```python
import matplotlib.pyplot as plt

# Data
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Create a scatter plot
plt.scatter(x, y, color='blue')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Test Scores')
plt.title('Age vs Test Scores')

# Show the plot
plt.show()
```

### Customizations
- Point Size:
- Point Colors:
- Adding Multiple Data Sets:
- Transparency:

```python
plt.scatter(x, y, s=10)  # Size of points (default is small)
```

```python
colors = ['red', 'blue', 'green', 'orange', 'purple','black','white','yellow','silver','pink']
plt.scatter(x, y, c=colors)
```

```python
x2 = [3, 6, 9]
y2 = [80, 85, 90]
plt.scatter(x, y, color='blue', label='Group 1')
plt.scatter(x2, y2, color='red', label='Group 2')
plt.legend()
```

```python
plt.scatter(x, y, alpha=0.5)  # Make points semi-transparent
```

## When to Use a Histogram?
- To show the shape of the data distribution (e.g., normal, skewed).
- To analyze frequency or density of data.
- To understand how data is spread (e.g., range, outliers).

```python
import matplotlib.pyplot as plt

# Data
data = [12, 15, 20, 21, 22, 25, 25, 27, 30, 31, 35, 36, 40, 42, 45, 50]

# Create a histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Show the plot
plt.show()
```

### Customizations
- Change the Number of Bins:
- Add Gridlines:
- Density Plot: Instead of frequency, show probability:
- Color and Styling:

```python
plt.hist(data, bins=10)  # More detailed histogram
```

```python
plt.grid(axis='y')  # Makes it easier to read
```

```python
plt.hist(data, bins=5, density=True)
```

```python
plt.hist(data, bins=5, color='green', alpha=0.7)  # Semi-transparent
```

### Creating Subplots

```python
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Add data to each subplot
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 0].set_title('Line Plot')

axs[0, 1].scatter([1, 2, 3], [6, 5, 4])
axs[0, 1].set_title('Scatter Plot')

axs[1, 0].bar(['A', 'B', 'C'], [7, 8, 9])
axs[1, 0].set_title('Bar Chart')

axs[1, 1].hist([1, 2, 2, 3, 3, 3, 4])
axs[1, 1].set_title('Histogram')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

## Pie Charts
- Visualizes proportions.

```python
labels = ['A', 'B', 'C']
sizes = [10, 30, 20]
plt.pie(sizes, labels=labels, autopct='%1.f%%')
plt.title('Pie Chart')
plt.show()
```

```python
def custom_autopct(pct):
    if pct > 5:
        return f'{pct:.1f}%'  # Show percentage for slices > 5%
    else:
        return ''  # Hide percentage for small slices

plt.pie(sizes, labels=labels, autopct=custom_autopct)
plt.title('Customized Percentage Display')
plt.show()
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 1, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [20, 30, 25, 25]
explode = [0, 0.1, 0, 0]  # Explode the second slice (Bananas)

# Create a pie chart
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)

# Add title
plt.title('Exploded Pie Chart Example')

# Show the chart
plt.show()
```

### Box Plot using Matplotlib

```python
import matplotlib.pyplot as plt

# Data
data = [12, 15, 20, 21, 22, 25, 25, 27, 30, 31, 35, 36, 40, 42, 45, 50]

# Create a box plot
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add labels and title
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```

### Customizing the Box Plot
- Horizontal Box Plot:
- Multiple Data Sets:
- Show Outliers:
- Change Colors:
- Add Gridlines:

```python
plt.boxplot(data, vert=False)  # Horizontal box plot
plt.show()
```

```python
data1 = [12, 15, 20, 21, 22]
data2 = [30, 35, 40, 45, 50]
plt.boxplot([data1, data2], labels=['Set 1', 'Set 2'])
plt.show()
```

```python
plt.boxplot(data, flierprops=dict(marker='o', color='red', markersize=10))
plt.show()
```

```python
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='green'))
plt.show()
```

```python
plt.boxplot(data)
plt.grid(True)  # Adds gridlines for better readability
plt.show()
```
